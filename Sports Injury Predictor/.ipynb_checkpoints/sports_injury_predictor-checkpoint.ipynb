```python
import pandas as pd
import numpy as np
import unidecode
```

```python
#datasets

player_injury_data = pd.read_csv("player_injuries_impact.csv")
epl_match_data = pd.read_csv("premier-league-matches.csv")
player_stats_data = pd.read_csv("All_Players_1992-2025.csv")

injury_df = pd.DataFrame(player_injury_data)
match_df = pd.DataFrame(epl_match_data)
stats_df = pd.DataFrame(player_stats_data)
```

```python
#convert N.A values in injury_df to NaN for proper stats

injury_df = injury_df.replace(["N.A", "N.A.", "NA", "N/A", "n.a", "NaN", ""], np.nan)
```

```python
injury_df.isna().sum()
```

```python
#drop columns containing info on matches after injury because we do not need them to predict

cols_to_drop = [col for col in injury_df.columns if "after_injury" in col.lower()]
cols_to_drop += [col for col in injury_df.columns if "missed_match" in col.lower()]

injury_df = injury_df.drop(columns=cols_to_drop)
```

```python
#standardize column names for merging per best practice 

injury_df.columns = injury_df.columns.str.lower().str.replace(" ", "_")
stats_df.columns  = stats_df.columns.str.lower().str.replace(" ", "_")
match_df.columns  = match_df.columns.str.lower().str.replace(" ", "_")
```

```python
#use unidecode to transform strings in injury and stats df and create column names for joining

injury_df["player_name"] = (
    injury_df["name"]
        .str.lower()
        .str.strip()
        .apply(unidecode.unidecode)
)

stats_df["player_name"] = (
    stats_df["player"]
        .str.lower()
        .str.strip()
        .apply(unidecode.unidecode)
)
```

```python
#normalize season because injuries and stats have different formats

def normalize_season(season):
    season = str(season)
    if "/" in season:
        start, end = season.split("/")
        return f"{start}-{start[:2]}{end}"
    return season
```

```python
injury_df["season_year"] = injury_df["season"].apply(normalize_season)
stats_df["season_year"] = stats_df["season"].apply(normalize_season)
```

```python
#filter league column on stats df to only have premier league

stats_df = stats_df[stats_df["league"].str.lower().str.contains("premier")].reset_index(drop=True)
```

```python
stats_df.head()
```

```python
#find teams with name mismatches so we can create a mapping to normalize

injury_teams = set(injury_df["team_name"].str.lower().str.strip())
injury_teams
```

```python
stats_teams = set(stats_df["squad"].str.lower().str.strip())
stats_teams
```

```python
match_home_teams = set(match_df["home"].str.lower().str.strip())
match_home_teams
```

```python
missing = injury_teams - stats_teams
missing
```

```python
#create a mapping for manchester united and newcastle because of name mismatches on the datasets

team_mapping = {
    # Manchester United
    "man united": "manchester united",
    "man utd": "manchester united",
    "manchester utd": "manchester united",

    # Newcastle
    "newcastle": "newcastle united",
    "newcastle utd": "newcastle united",
}
```

```python
#normalize team names for merging 

injury_df["player_team"] = (
    injury_df["team_name"]
        .str.lower()
        .str.strip()
        .replace(team_mapping)
)

stats_df["player_team"] = (
    stats_df["squad"]
        .str.lower()
        .str.strip()
        .replace(team_mapping)
)
```

```python
injury_df.head()
```

```python
#make sure common rows in both columns are returned

set(injury_df["player_name"]) & set(stats_df["player_name"])
```

```python
set(injury_df["player_team"]) & set(stats_df["player_team"])
```

```python
set(injury_df["season_year"]) & set(stats_df["season_year"])
```

```python
#merge injury and stats dataframes using player_name, player_team, and season_year using a left join

merged_df = injury_df.merge(
    stats_df,
    on=["player_name", "player_team", "season_year"],
    how="left"
)
```

```python
#confirm merge works

merged_df.info()
merged_df.head()
```

```python
#normalize team names on match df

match_df["home_team"] = (
    match_df["home"]
        .str.lower()
        .str.strip()
        .replace(team_mapping)
)

match_df["away_team"] = (
    match_df["away"]
        .str.lower()
        .str.strip()
        .replace(team_mapping)
)
```

```python
# add a date-time version of date

match_df["match_date"] = pd.to_datetime(match_df["date"])
```

```python
#normalize season_end_year. football seasons are always written by the end year, e.g. 1993 end year means 1992-1993 season. so we can subtract one year and make a conversion

def convert_season_end_year(year):
    year = int(year)
    start = year - 1
    end = year
    return f"{start}-{end}"
```

```python
match_df["season_year"] = match_df["season_end_year"].apply(convert_season_end_year)
```

```python
#next, filter by season year and teams to make sure we only have teams and years in the merged df

valid_teams = set(merged_df["player_team"].unique())
valid_seasons = set(merged_df["season_year"].unique())

match_df = match_df[match_df["season_year"].isin(valid_seasons)]
match_df = match_df[
    (match_df["home_team"].isin(valid_teams)) |
    (match_df["away_team"].isin(valid_teams))
]
```
